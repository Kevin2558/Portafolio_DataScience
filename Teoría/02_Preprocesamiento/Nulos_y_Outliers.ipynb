# Nulos y Outliers

Uno de los primeros pasos fundamentales en cualquier pipeline de análisis de datos es la identificación y tratamiento de valores nulos y outliers. Estos elementos pueden distorsionar el análisis estadístico, afectar los modelos predictivos y generar resultados poco confiables.

Separemos la variables Survived entre sus dos clases, sobrevivió o no.

```python
survived = df[df['Survived'] == 1]['Embarked']
not_survived = df[df['Survived'] == 0]['Embarked']
```

```python
import matplotlib.pyplot as plt
```

Lo siguiente se realizará para visualizar los sobrevivientes por cabina.

```python
surv_counts = survived.value_counts()
nonsurv_counts = not_survived.value_counts()

all_categories = sorted(list(set(surv_counts.index) | set(nonsurv_counts.index)))
surv_counts = surv_counts.reindex(all_categories, fill_value=0)
nonsurv_counts = nonsurv_counts.reindex(all_categories, fill_value=0)

width = 0.4
x=range(len(all_categories))

plt.bar(x, surv_counts.values, width=width, alpha=0.7, label='Survived')
plt.bar([i + width for i in x], nonsurv_counts.values, width=width, alpha=0.7, label='Not Survived')

plt.xticks([i+width/2 for i in x], all_categories)
plt.title('Comparacion de ciudades (Supervivencia)')
plt.xlabel('Cabin')
plt.ylabel('Count')
plt.legend()
plt.show()
```

Como podemos ver no es clara una relacion entre la ciudad y si se sobrevive o no por lo que se podria imputar por la que mas se repite.

Ahora, comencemos con el analisis y el tratamiento de los outliers.

Sabemos que el indice intercuartilico (IQR) es mas estable cuando hay una mayor cantidad de outliers, debido a que mientras mas valores desviados existan la media y la desviación estándar se verán más afectados. En el caso de que el histograma es mas o menos normal, podemos usar z-score sin problema.

Calculemos el IQR para la edad:

```python
Q1 = df['Age'].quantile(0.25)
Q3 = df['Age'].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers_iqr = df[(df['Age'] < lower_bound) | (df['Age'] > upper_bound)]

print('Numero de outliers (metodo IQR):', len(outliers_iqr))
```

```python
import numpy as np
```

Lo siguiente que haremos erá filtrar todas las filas de df tal que su z_scores sea mayor que 3.

```python
mean_age = df['Age'].mean()
std_age = df['Age'].std()

z_scores = np.abs((df['Age'] - mean_age) / std_age)

outliers_zscore = df[z_scores > 3]
```

```python
print('Numero de outliers (metodo z-score):', len(outliers_zscore))
```

Como podemos notar para el metodo iqr son 66 pero para el metodo z-score son 7, esto debido a que en el caso de la edad la media y la desviacion estandar se veian afectados. Por lo que es recomendable usar iqr si el histograma no es tan normal.

La siguiente libreria calcula automaticamente el zscore.

```python
from scipy.stats import zscore
```

```python
z = zscore(df['Age'])
```

```python
Q1 = df_houses['price'].quantile(0.25)
Q3 = df_houses['price'].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
```

```python
outliers_iqr['price'].sort_values(ascending=False)
```
