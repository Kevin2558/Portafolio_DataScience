# LDA (Análisis Discriminante Lineal)

Técnica supervisada que encuentra una proyección de los datos que maximiza la separabilidad entre clases. A diferencia de PCA, LDA utiliza etiquetas de clase.

Casos de uso recomendados:
- Clasificación supervisada.
- Visualización de datos etiquetados.
- Reducción de dimensionalidad con retención de discriminación entre clases.


Veamos LDA aplicado al dataset de dígitos.

```python
import matplotlib.pyplot as plt
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.datasets import load_digits
from sklearn.preprocessing import StandardScaler

# Cargamos los datos
digits = load_digits()
X = digits.data
y = digits.target

# Estandarizamos las variables
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

Aplicamos LDA.

```python
lda = LinearDiscriminantAnalysis(n_components=2)
X_lda = lda.fit_transform(X_scaled, y)
```

Visualizamos los datos proyectados.

```python
plt.figure(figsize=(8,6))
scatter = plt.scatter(X_lda[:, 0], X_lda[:, 1], c=y, cmap='tab10', alpha=0.7)
plt.legend(*scatter.legend_elements(), title="Dígitos")
plt.title("Proyección con LDA")
plt.xlabel("Componente 1")
plt.ylabel("Componente 2")
plt.grid(True)
plt.show()
```
