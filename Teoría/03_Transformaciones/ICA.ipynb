# ICA (Análisis de Componentes Independientes)

Busca componentes estadísticamente independientes maximizando la no-gaussianidad. Se diferencia de PCA porque no solo considera varianza, sino independencia.

Casos de uso recomendados:
- Separación de señales (por ejemplo, señales EEG, mezcla de audios o imágenes).
- Análisis en neurociencia y bioseñales.
- Preprocesamiento cuando se sospecha que las fuentes subyacentes son independientes.

Veamos ICA aplicado al dataset de dígitos.

```python
import matplotlib.pyplot as plt
from sklearn.decomposition import FastICA
from sklearn.datasets import load_digits
from sklearn.preprocessing import StandardScaler

# Cargamos los datos
digits = load_digits()
X = digits.data
y = digits.target

# Estandarizamos las variables
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

Aplicamos ICA

```python
ica = FastICA(n_components=2, random_state=0)
X_ica = ica.fit_transform(X_scaled)
```

Visualizamos los datos proyectados

```python
plt.figure(figsize=(8,6))
scatter = plt.scatter(X_ica[:, 0], X_ica[:, 1], c=y, cmap='tab10', alpha=0.7)
plt.legend(*scatter.legend_elements(), title="Dígitos")
plt.title("Proyección con ICA")
plt.xlabel("Componente 1")
plt.ylabel("Componente 2")
plt.grid(True)
plt.show()
```
