# UMAP (Proyección y Aproximación de Variedades Uniformes)

Proyección no lineal que modela relaciones topológicas entre puntos usando grafos. Es más rápida que t-SNE y mantiene mejor la estructura global y local.

Casos de uso recomendados:
- Visualización exploratoria (incluso en datasets grandes).
- Reducción de dimensionalidad antes de clustering o clasificación.
- Análisis donde se necesita preservar estructura local y global.


Veamos UMAP aplicado al dataset de dígitos.

```python
import matplotlib.pyplot as plt
import umap.umap_ as umap
from sklearn.datasets import load_digits
from sklearn.preprocessing import StandardScaler

# Cargamos los datos
digits = load_digits()
X = digits.data
y = digits.target

# Estandarizamos los datos
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

Aplicamos UMAP.

```python
reducer = umap.UMAP(n_components=2, random_state=42)
X_umap = reducer.fit_transform(X_scaled)
```

Visualizamos los datos proyectados.

```python
plt.figure(figsize=(8,6))
scatter = plt.scatter(X_umap[:, 0], X_umap[:, 1], c=y, cmap='tab10', alpha=0.7)
plt.legend(*scatter.legend_elements(), title="Dígitos")
plt.title("Proyección con UMAP")
plt.xlabel("Componente 1")
plt.ylabel("Componente 2")
plt.grid(True)
plt.show()
```
