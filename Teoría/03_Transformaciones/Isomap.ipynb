# Isomap

Técnica no lineal que preserva las distancias geodésicas (en el manifold). Usa un grafo de vecinos para mantener la geometría intrínseca del conjunto de datos.

Casos de uso recomendados:
- Cuando se sospecha que los datos viven en una variedad (manifold) no lineal.
- Visualización de datos de alta dimensión con estructura curva.
- No recomendado en presencia de mucho ruido (sensible).

Veamos Isomap aplicado al dataset de dígitos.

```python
import matplotlib.pyplot as plt
from sklearn.manifold import Isomap
from sklearn.datasets import load_digits
from sklearn.preprocessing import StandardScaler

# Cargamos los datos
digits = load_digits()
X = digits.data
y = digits.target

# Estandarizamos los datos
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

Aplicamos Isomap.

```python
isomap = Isomap(n_components=2, n_neighbors=5)
X_iso = isomap.fit_transform(X_scaled)
```

Visualizamos los datos proyectados.

```python
plt.figure(figsize=(8,6))
scatter = plt.scatter(X_iso[:, 0], X_iso[:, 1], c=y, cmap='tab10', alpha=0.7)
plt.legend(*scatter.legend_elements(), title="Dígitos")
plt.title("Proyección con Isomap")
plt.xlabel("Componente 1")
plt.ylabel("Componente 2")
plt.grid(True)
plt.show()
```
