# Kernel PCA (Análisis de Componentes Principales con kernel)

Es una extensión no lineal de PCA que permite proyectar los datos a un espacio de mayor dimensión usando una función kernel (RBF, polinómico, etc.), para capturar estructuras no lineales.

Casos de uso recomendados:
- Cuando los datos no son linealmente separables.
- Para detección de patrones no lineales en datos complejos.
- Cuando se quiera hacer reducción de dimensionalidad con estructuras no lineales preservadas.


Veamos Kernel PCA aplicado al dataset de dígitos.

```python
import matplotlib.pyplot as plt
from sklearn.decomposition import KernelPCA
from sklearn.datasets import load_digits
from sklearn.preprocessing import StandardScaler

# Cargamos los datos
digits = load_digits()
X = digits.data
y = digits.target

# Estandarizamos las variables antes de la reduccion
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

Aplicamos Kernel PCA con kernel RBF

```python
kpca = KernelPCA(n_components=2, kernel='rbf', gamma=0.04)
X_kpca = kpca.fit_transform(X_scaled)
```

Visualizamos los datos proyectados

```python
plt.figure(figsize=(8,6))
scatter = plt.scatter(X_kpca[:, 0], X_kpca[:, 1], c=y, cmap='tab10', alpha=0.7)
plt.legend(*scatter.legend_elements(), title="Dígitos")
plt.title("Proyección con Kernel PCA")
plt.xlabel("Componente 1")
plt.ylabel("Componente 2")
plt.grid(True)
plt.show()
```
